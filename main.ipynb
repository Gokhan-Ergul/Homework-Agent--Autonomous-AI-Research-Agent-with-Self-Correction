```python
long_research_query = HumanMessage(
content= """
Write a detailed report on the future of renewable energy sources (solar, wind, geothermal) in Türkiye.

This report must be at least 5 pages (approximately 2000 words) long.

The report should include the following sections:

1. **Introduction:** Türkiye's current energy profile and its need for renewable energy.

2. **Solar Energy Potential:** Current installed capacity, geographical advantages, and solar power plant projects.

3. **Wind Energy Potential:** The most productive regions for wind power plants, installed capacity, and challenges.

4. **Geothermal Energy Potential:** Türkiye's location in the geothermal belt and geothermal power plant projects.

5. **Challenges and Incentives:** Legal regulations, investment costs, and government incentives.

6. **Conclusion:** 2030 and An overall assessment in line with the 2050 goals.

Please use a web search to analyze each section in depth, and support your claims with statistics and current examples.
""")


start_time = time.time()
response = app.invoke({'messages': [long_research_query]})
    
end_time = time.time()

duration_seconds = end_time - start_time
print(f"Time elapsed: {duration_seconds}")
```
