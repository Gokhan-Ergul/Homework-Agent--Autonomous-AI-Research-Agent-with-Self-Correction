```python
new_complex_query = HumanMessage(
    content="""
Write a 400-word academic summary on the evaluation and application of memory in LLM-based agents.

1. Start by defining what 'memory' means for an LLM agent and why it is crucial for applications like role-playing.
2. Compare and contrast 'direct evaluation' and 'indirect evaluation' (like success rate or exploration degree in games) for an agent's memory.
3. Discuss the main challenges or biases in evaluating agent memory, based on recent academic surveys.
4. Conclude by listing at least three major application areas for memory-enhanced agents.

Use a web search to find current academic papers or surveys on these topics.
"""
)

print("--- STARTING AGENT INVOCATION ---")
start_time = time.time()

response = app.invoke({'messages': [new_complex_query]})

end_time = time.time()
duration_seconds = end_time - start_time
print(f"--- INVOCATION FINISHED IN {duration_seconds:.2f} SECONDS ---")
```
